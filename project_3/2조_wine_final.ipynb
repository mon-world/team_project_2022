```python
def pltplt(X_train, X_test, y_train, y_test):

    x = []

    #LogisticRegression
    lr_model = LogisticRegression()
    lr_model.fit(X_train, y_train)
    lr_pred = lr_model.predict(X_test)
    # print(classification_report(y_test, lr_pred))
    print("LogisticRegression 정확도: ", metrics.accuracy_score(y_test, lr_pred) )
    x.append(metrics.accuracy_score(y_test, lr_pred))

    #KNN
    knn_model = KNeighborsClassifier(5)
    knn_model.fit(X_train, y_train)
    knn_pred = knn_model.predict(X_test)
    # print(classification_report(y_test, knn_pred) )
    print("KNN 정확도: ", metrics.accuracy_score(y_test, knn_pred) )
    x.append(metrics.accuracy_score(y_test, knn_pred))

    #SVC
    best_parameters = {'C': 100, 'gamma': 1}
    svc = SVC(**best_parameters)
    svc.fit(X_train, y_train)
    svc_pred = svc.predict(X_test)
    # print(classification_report(y_test, svc_pred) )
    print("SVC 정확도: ", metrics.accuracy_score(y_test, svc_pred) )
    x.append(metrics.accuracy_score(y_test, svc_pred)) 

    # DecisionTreeClassifier
    tr_model = tree.DecisionTreeClassifier()
    tr_model.fit(X_train, y_train)
    tr_pred = tr_model.predict(X_test)
    # print(classification_report(y_test, tr_pred))
    print("DecisionTreeClassifier 정확도: ", metrics.accuracy_score(y_test, tr_pred) )
    x.append(metrics.accuracy_score(y_test, tr_pred)) 

    # RandomForestClassifier
    wine_rfc = RandomForestClassifier(n_estimators=100)
    wine_rfc.fit(X_train, y_train)
    wine_rfc_pred = wine_rfc.predict(X_test)
    # print( classification_report(y_test, wine_rfc_pred) )
    print( "RandomForestClassifier 정확도: ", metrics.accuracy_score(y_test, wine_rfc_pred) )
    x.append(metrics.accuracy_score(y_test, wine_rfc_pred)) 

    return x
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()  # set_theme() 메서드의 별칭. style 매개변수의 기본값은 'darkgrid'

start, end, step = 50, 1000, 50

# OOB 점수를 담을 빈 리스트
oob_scores = []
model_rf = RandomForestClassifier(n_estimators=50, warm_start=True,
                               oob_score=True, n_jobs=-1, random_state=2)

# 모델 훈련과 oob_score_ 값을 리스트에 추가
model_rf.fit(X_train, y_train)
oob_scores.append(model_rf.oob_score_)

# 50부터 시작해서 트리 개수를 담을 리스트 생성
est = start
estimators = [est]

# 50개씩 트리를 추가하는 반복문
n = end // start
for i in range(n):
  est += step
  estimators.append(est)
  model_rf.set_params(n_estimators=est)
  model_rf.fit(X_train, y_train)
  oob_scores.append(model_rf.oob_score_)
```

```python
plt.figure( figsize=(10, 4) )
plt.plot(estimators, oob_scores)

plt.xlabel('Number of Trees')
plt.ylabel('oob_score_')
plt.title('Random Forest Warm Start', fontsize='16')
plt.show()
```
