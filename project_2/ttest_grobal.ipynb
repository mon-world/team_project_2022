```python
def deff_renew(df, country, start, end):
    country = df.loc[f'{country}', start : end].values
    country_d = []
    for i in range(len(country)-1):
        country_d.append(country[i+1] - country[i])

    return country_d
```

```python
g_seven = ['United States', 'Japan', 'Germany', 'United Kingdom', 'France', 'Italy', 'Canada']
```

```python
start = '2016.0'
end = '2020.0'

kor = deff_renew(df, 'South Korea', start, end)
world = deff_renew(df, 'World',  start, end)
g7 = deff_renew(df, 'G7',  start, end)
us = deff_renew(df, 'United States', start, end)
jp = deff_renew(df, 'Japan', start, end)
gr = deff_renew(df, 'Germany', start, end)
uk = deff_renew(df, 'United Kingdom', start, end)
fr = deff_renew(df, 'France', start, end)
it = deff_renew(df, 'Italy', start, end)
cn = deff_renew(df, 'Canada', start, end)
```
