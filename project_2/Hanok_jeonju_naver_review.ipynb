```python
# 네이버맵 전주한옥마을 리뷰창
url = 'https://pcmap.place.naver.com/place/13208779/review/visitor?from=map&fromPanelNum=1&ts=1658451376983'

print(url)
```

```python
driver = webdriver.Chrome('chromedriver', chrome_options=chrome_options)

driver.get(url) 
time.sleep(3)
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") 
time.sleep(3)

btn = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/div/div[7]/div[2]/div[4]/h2/a')
btn.click()
time.sleep(3 + np.random.rand())

driver.save_screenshot('save01.png')
```

```python
# 시간이 많거나, 딴짓할게 있을 때 하기
# 시작 시간
start_time = time.time()
i = 1
flag = [0]

# 크롤링
while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")    # 스크롤 하는 함수 0이 꼭대기
    time.sleep(3 + np.random.rand())

    # while문 멈출 고삐
    if i % 5 == 0:
        check = driver.execute_script("return document.body.scrollHeight")      # 바 내렸을 때 높이값
        flag.append(check)

        # flag에 값이 같으면 탈출, 아니면 flag 초기화
        if flag[0] == flag[1]:
            print('걸린 시간 : ', time.time() - start_time)  
            break
        else:
            print(f'{i}회 확장 동안 새로 생긴 스크롤바 높이 : ', flag[1] - flag[0])
            flag = [check]
            print('걸린 시간 : ', time.time() - start_time)
            time.sleep(10 + 10*np.random.rand())
        
    # 더보기를 클릭해야 새로운 리뷰가 열린다
    driver.find_element(By.XPATH, '/html/body/div[3]/div/div/div/div[7]/div[2]/div[4]/div[2]/a').click()
    time.sleep(3 + np.random.rand())

    # 혹시 모를 끊김을 대비하자! : 중간 저장
    if i % 10 == 0:
        try:
            web_temp = driver.page_source
            soup_temp = BeautifulSoup(web_temp, 'html.parser')
            contents = soup.select('li._3l2Wz span.utrsf')

            stars, ago,  = [], []

            for content in contents:
                # print(content.text) 
                if content.text.startswith('최근'):
                    ago.append(content.text[6:])
                elif content.text.endswith('요'):
                    stars.append(content.text)
                elif content.text.endswith('별점'):
                    stars.append(content.text)
                else:
                    pass

            df_temp = pd.DataFrame({'star' : stars,
                                    'ago' : ago})
            
            df_temp.to_csv(drive_path + f'중간저장/reviews_naver_{i // 10}.csv', encoding='utf-8')
        except:
            pass

    i += 1

web = driver.page_source
soup = BeautifulSoup(web, 'html.parser')
```
