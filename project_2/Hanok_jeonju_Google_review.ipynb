```python
# 시간이 많거나, 딴짓할게 있을 때 하기
# 시작 시간
start_time = time.time()

for i in range(2570):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") # 스크롤 하는 함수 0이 꼭대기
    time.sleep(1 + np.random.rand())

    driver.find_element(By.TAG_NAME, "body").send_keys(Keys.PAGE_UP)
    time.sleep(1 + np.random.rand())

    # 진행 상황 확인
    if i % 100 == 0:
        print(f'{i} 번째 화면 확장중')

    # 혹시 모를 끊김을 대비하자!
    web_temp = driver.page_source
    soup = BeautifulSoup(web_temp, 'html.parser')

web = driver.page_source
soup = BeautifulSoup(web, 'html.parser')

# 걸린 시간 계산
print("끝! 걸린 시간 : ", time.time() - start_time)
```

```python
driver.save_screenshot('sb00.png')
```

```python
driver.close()
driver.quit()
```

```python
# 별점, 글 쓴 시점, 리뷰 내용
stars = [int(star.text.strip()[0]) for star in soup.find_all('div', class_='GDWaad')]
times = [time.text.strip()[:-16] for time in soup.select('div.jUkSGf span.iUtr1.CQYfx')]
reviews = [review.text.strip() for review in soup.select('div.STQFb.eoY5cb')]

# 길이 확인
print(len(stars), len(times), len(reviews))
```

```python
# df로 만들기
df = pd.DataFrame({'star' : stars,
                   'time' : times,
                   'review_content' : reviews},
                  index=list(range(len(stars))))
```

```python
time_temp = list(set(times))
time_temp2 = sorted([t[::-1] for t in time_temp])
time_temp3 = [t[::-1] for t in time_temp2]
print(time_temp3)
```
