### 1.   구글 리뷰

```python
# 구글맵 전주한옥마을 리뷰창
url = 'https://www.google.co.kr/travel/entity/key/ChgIjonCtrzDr_qOARoLL2cvMXRjX3pjbHoQBA/reviews'

print(url)
```

```python
driver = webdriver.Chrome('chromedriver', chrome_options=chrome_options)

driver.get(url) 
time.sleep(3)
driver.save_screenshot('sb00.png')
```

```python
# 시간이 많거나, 딴짓할게 있을 때 하기
# 시작 시간
start_time = time.time()
i = 1
flag = [0]

# 크롤링
while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")    # 스크롤 하는 함수 0이 꼭대기
    time.sleep(3 + np.random.rand())

    # while문 멈출 고삐
    if i % 500 == 0:
        check = driver.execute_script("return document.body.scrollHeight")      # 바 내렸을 때 높이값
        flag.append(check)

        # flag에 값이 같으면 탈출, 아니면 flag 초기화
        if flag[0] == flag[1]:
            print("끝! 걸린 시간 : ", (time.time() - start_time) // 3600, '시간 ', (time.time() - start_time) % 3600, '분')  
            break
        else:
            print(f'{i}회 확장 동안 새로 생긴 스크롤바 높이 : ', flag[1] - flag[0])
            flag = [check]
            print((time.time() - start_time) // 3600, '시간 ', (time.time() - start_time) % 3600, '분 경과했습니다.')
            time.sleep(20 + 30*np.random.rand())
        
    # 스크롤바를 올려줘야 로딩이 된다.
    driver.find_element(By.TAG_NAME, "body").send_keys(Keys.PAGE_UP)
    time.sleep(3 + np.random.rand())

    # 혹시 모를 끊김을 대비하자! : 중간 저장
    if i % 1000 == 0:
        try:
            web_temp = driver.page_source
            soup_temp = BeautifulSoup(web_temp, 'html.parser')

            stars = [int(star.text.strip()[0]) for star in soup_temp.find_all('div', class_='GDWaad')]
            p_times = [p_time.text.strip()[:-16] for p_time in soup_temp.select('div.jUkSGf span.iUtr1.CQYfx')]
            reviews = [review.text.strip() for review in soup_temp.select('div.STQFb.eoY5cb')]

            df_temp = pd.DataFrame({'star' : stars,
                                    'ago' : p_times,
                                    'review_content' : reviews})
            
            df_temp.to_csv(drive_path + f'중간저장/reviews_{i // 3000}.csv', encoding='utf-8')
        except:
            pass

    i += 1

web = driver.page_source
soup = BeautifulSoup(web, 'html.parser')
```

```python
# 별점, 글 쓴 시점, 리뷰 내용
stars = [int(star.text.strip()[0]) for star in soup.find_all('div', class_='GDWaad')]
p_times = [p_time.text.strip()[:-16] for p_time in soup.select('div.jUkSGf span.iUtr1.CQYfx')]
reviews = [review.text.strip() for review in soup.select('div.STQFb.eoY5cb')]

# 길이 확인
print(len(stars), len(p_times), len(reviews))
```

```python
# df로 만들기
df = pd.DataFrame({'star' : stars,
                   'ago' : p_times,
                   'review_content' : reviews})
```
